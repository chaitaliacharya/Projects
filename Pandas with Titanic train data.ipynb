```python
def pclass_name(x):
    if x == 1:
        x = '1st Class'
    elif x == 2:
        x = '2nd Class'
    elif x == 3:
        x = '3rd Class'
    return x
```

```python
df3['Pclass'] = df3['Pclass'].apply(lambda x: pclass_name(x))
```

```python
df3
```

```python
def height_convert(x):
    x = (x/100)**2
    return x
df3['height'] = df3['height'].apply(lambda x: height_convert(x))
    
```

```python
df3
```

```python
df3['Bmi'] = df3['weight']/df3['height']
```
